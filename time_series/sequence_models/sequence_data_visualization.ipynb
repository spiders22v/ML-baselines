```python
# 라인 차트 데이터 생성

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 가상의 주식 데이터 생성 (변동폭이 큰 데이터)
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=200)
# 초기 가격 설정 및 변동폭 크게 조정
prices = 100 + np.cumsum(np.random.randn(200) * 10)

# 데이터프레임 생성
stock_data = pd.DataFrame(data={'Date': dates, 'Price': prices})

# 데이터프레임에서 Date를 인덱스로 설정
stock_data.set_index('Date', inplace=True)

# 주식 데이터 시각화
plt.figure(figsize=(14, 7))
plt.plot(stock_data.index, stock_data['Price'], label='Stock Price', color='blue')
plt.title('Stock Price Over Time with High Volatility')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 히트맵 예제 데이터 생성

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

# 10x10 크기의 예제 데이터 생성 (정규분포를 따름)
data = np.random.randn(10, 10)

# 데이터프레임 생성
df = pd.DataFrame(data, columns=[f'Column {i+1}' for i in range(10)], index=[f'Row {i+1}' for i in range(10)])

# 히트맵 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(df, annot=True, fmt=".2f", cmap='coolwarm', linewidths=.5, linecolor='black')
plt.title('Heatmap of Example Data')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```

```python
# 시즌 플롯 예제 데이터 생성
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 월별 기온 데이터 생성
np.random.seed(42)
months = pd.date_range(start='2018-01-01', periods=36, freq='M')
# 기온 데이터 생성 (연간 주기를 갖는 사인 곡선에 노이즈 추가)
temps = 10 + 10 * np.sin(2 * np.pi * months.month / 12) + np.random.normal(0, 2, len(months))

# 데이터프레임 생성
df = pd.DataFrame(data={'Date': months, 'Temperature': temps})
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# 시즌 플롯 시각화
plt.figure(figsize=(14, 8))
sns.lineplot(x='Month', y='Temperature', hue='Year', data=df, marker='o', palette='tab10')

plt.title('Seasonal Plot of Monthly Temperature Changes')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.legend(title='Year')
plt.grid(True)
plt.show()
```

```python
# 단순 이동 평균 예제 데이터 생성

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 월별 기온 데이터 생성 (36개월)
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', periods=36, freq='M')
# 기온 데이터 생성 (연간 주기를 갖는 사인 곡선에 노이즈 추가)
temps = 10 + 10 * np.sin(2 * np.pi * dates.month / 12) + np.random.normal(0, 2, len(dates))

# 데이터프레임 생성
df = pd.DataFrame(data={'Date': dates, 'Temperature': temps})

# 3개월 이동 평균 계산
df['SMA_3'] = df['Temperature'].rolling(window=3).mean()

# 데이터프레임 확인
print(df.head())

# 원본 데이터와 이동 평균 시각화
plt.figure(figsize=(14, 8))
plt.plot(df['Date'], df['Temperature'], marker='o', label='Original Temperature')
plt.plot(df['Date'], df['SMA_3'], color='red', marker='o', label='3-Month Simple Moving Average')
plt.title('Monthly Temperature with 3-Month Simple Moving Average')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 가중 이동 평균 시각화

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 월별 기온 데이터 생성 (36개월)
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', periods=36, freq='M')
# 기온 데이터 생성 (연간 주기를 갖는 사인 곡선에 노이즈 추가)
temps = 10 + 10 * np.sin(2 * np.pi * dates.month / 12) + np.random.normal(0, 2, len(dates))

# 데이터프레임 생성
df = pd.DataFrame(data={'Date': dates, 'Temperature': temps})

# 가중 이동 평균 계산 함수
def weighted_moving_average(values, window):
    weights = np.arange(1, window + 1)
    wma = np.convolve(values, weights[::-1], 'valid') / weights.sum()
    return wma

# 3개월 가중 이동 평균 계산
window_size = 3
df['WMA_3'] = np.nan
df['WMA_3'][window_size - 1:] = weighted_moving_average(df['Temperature'], window_size)

# 데이터프레임 확인
print(df.head(window_size + 5))

# 원본 데이터와 가중 이동 평균 시각화
plt.figure(figsize=(14, 8))
plt.plot(df['Date'], df['Temperature'], marker='o', label='Original Temperature')
plt.plot(df['Date'], df['WMA_3'], color='red', marker='o', label='3-Month Weighted Moving Average')
plt.title('Monthly Temperature with 3-Month Weighted Moving Average')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 월별 기온 데이터 생성 (36개월)
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', periods=36, freq='M')
# 기온 데이터 생성 (연간 주기를 갖는 사인 곡선에 노이즈 추가)
temps = 10 + 10 * np.sin(2 * np.pi * dates.month / 12) + np.random.normal(0, 2, len(dates))

# 데이터프레임 생성
df = pd.DataFrame(data={'Date': dates, 'Temperature': temps})

# 지수 이동 평균 계산
df['EMA_3'] = df['Temperature'].ewm(span=3, adjust=False).mean()

# 데이터프레임 확인
print(df.head())

# 원본 데이터와 지수 이동 평균 시각화
plt.figure(figsize=(14, 8))
plt.plot(df['Date'], df['Temperature'], marker='o', label='Original Temperature')
plt.plot(df['Date'], df['EMA_3'], color='red', marker='o', label='3-Month Exponential Moving Average')
plt.title('Monthly Temperature with 3-Month Exponential Moving Average')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 단순 이동 평균, 가중 이동 평균, 지수 이동 평균을 같은 그래프에서 시각화

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 월별 기온 데이터 생성 (36개월)
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', periods=36, freq='M')
# 기온 데이터 생성 (연간 주기를 갖는 사인 곡선에 노이즈 추가)
temps = 10 + 10 * np.sin(2 * np.pi * dates.month / 12) + np.random.normal(0, 2, len(dates))

# 데이터프레임 생성
df = pd.DataFrame(data={'Date': dates, 'Temperature': temps})

# 단순 이동 평균 계산
df['SMA_3'] = df['Temperature'].rolling(window=3).mean()

# 가중 이동 평균 계산 함수
def weighted_moving_average(values, window):
    weights = np.arange(1, window + 1)
    wma = np.convolve(values, weights[::-1], 'valid') / weights.sum()
    return wma

# 3개월 가중 이동 평균 계산
window_size = 3
df['WMA_3'] = np.nan
df['WMA_3'][window_size - 1:] = weighted_moving_average(df['Temperature'], window_size)

# 지수 이동 평균 계산
df['EMA_3'] = df['Temperature'].ewm(span=3, adjust=False).mean()

# 데이터프레임 확인
print(df.head(window_size + 5))

# 원본 데이터와 이동 평균들 시각화
plt.figure(figsize=(14, 8))
plt.plot(df['Date'], df['Temperature'], color='black', marker='o', label='Original Temperature')
plt.plot(df['Date'], df['SMA_3'], color='red', marker='o', label='3-Month Simple Moving Average')
plt.plot(df['Date'], df['WMA_3'], color='green', marker='o', label='3-Month Weighted Moving Average')
plt.plot(df['Date'], df['EMA_3'], color='blue', marker='o', label='3-Month Exponential Moving Average')
plt.title('Monthly Temperature with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```
